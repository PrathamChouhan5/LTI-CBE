```python
x = df.iloc[:,:-1]
```

```python
y = df['medv']
```

```python
# Linear, Non-linear, Ensemble models
```

```python
# Linear regression --> OLS [ordinary least squares] --> least squared error(residuaL)
# Line of best fit
# co-efficient/weight
# Intercept
# expected(y) ==> b0+b1*x1+b2*x2+b3*x3+.........b13*x13
# expected(y) --> 1.5+(0.66)*4
# Assumption 1 --> y & all x vars should have some correlation
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
xtrain[:2]
```

```python
ytest[:2]
```

```python
xtrain.shape
```

```python
ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(xtrain,ytrain)
```

```python
# stats model
```

```python
dict(zip(xtrain.columns,model.coef_))
```

```python
model.intercept_
```

```python
xtest[:1]
```

```python
#e(price) = 37+(3.63*6.51)+(-0.55*10.29)
```

```python
pred = model.predict(xtest)
```

```python
pred[:5]
```

```python
ytest[:5]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(ytest,pred)
```

```python
import numpy as np
```

```python
np.sqrt(mean_squared_error(ytest,pred))
```

```python
model.predict(xtest[:1])
```

```python
error = pd.DataFrame({'actual':ytest,'predicted':pred})
```

```python
error.reset_index(drop=True, inplace=True)
```

```python
plt.plot(error)
```

```python
# Non-linear  algorithms --> Tree, SVM, KNN
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(min_samples_leaf=5, min_samples_split=5, max_depth=5)
dt_model = dt.fit(xtrain,ytrain)
```

```python
from dtreeviz import model

viz_model = model(
    dt_model,
    X_train=xtrain,
    y_train=ytrain,
    feature_names=x.columns.to_list(),
    target_name='medv'
)
viz_model.view()
```

```python
dtpred= dt_model.predict(xtest)
```

```python
np.sqrt(mean_squared_error(ytest,dtpred))
```

```python
np.sqrt(mean_squared_error(ytest,pred))
```

```python
# Ensemble models --> Bagging(Random forest), Boosting
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100)
```

```python
rfmodel = rf.fit(xtrain,ytrain)
```

```python
predrf = rfmodel.predict(xtest)
```

```python
np.sqrt(mean_squared_error(ytest,predrf))
```
